# `mammos_spindynamics` quickstart

`mammos_spindynamics` provides temperature-dependent spontaneous magnetization values.

- `db` contains pre-computed materials
- `uppasd` provides a Python interface to [UppASD](https://github.com/UppASD/UppASD)

````python
import mammos_spindynamics
````

## Querying the database – `mammos_spindynamics.db`

Use the following function to get a list of all available materials:

````python
mammos_spindynamics.db.find_materials()
````

Use the following function to get an object that contains spontaneous magnetization `Ms` at temperatures `T`:

````python
results_spindynamics = mammos_spindynamics.db.get_spontaneous_magnetization("Co2Fe2H4")
````

The result object provides a function to plot the data for visual inspection:

````python
results_spindynamics.plot()
````

We can access the `Ms` and `T` attributes and get `mammos_entity.Entity` objects:

````python
results_spindynamics.Ms
````

````python
results_spindynamics.T
````

To work with the data we can also get it as `pandas.DataFrame`. The dataframe only contains the values (not units).

````python
results_spindynamics.dataframe
````

## Running spindynamics simulations – `mammos_spindynamics.uppasd`

This subpackage provides an Python interface to UppASD. UppASD has to be installed separately and the executable must be in your `PATH`.

In this notebook we will show how to run a single simulation. We assume that we have all required input files. [...] will explain both creating inputs as well as analysing the resulting data in more detail.

We assume that we have the following files in the current working directory: `inpsd.dat`, `jfile`, `posfile`, `momfile`.

`inpsd.dat` references the other three as:
```
...
posfile = ./posfile
momfile = ./momfile
exchange = ./jfile
...
```

Furthermore, it has a placeholder `TEMP` for `ip_temp` and `temp`:
```
...
ip_mode TEMP
...
temp TEMP
...
```

First, we need to create a simulation object:

````python
sim = mammos_spindynamics.uppasd.Simulation.from_inpsd(
    "inpsd.dat", posfile="posfile", momfile="momfile", exchange="jfile"
)
````

We can now use this object to run a simulation at a specific temperature. The temperature value must be in Kelvin:

````python
result = sim.run(T=10, out="my-first-simulation")
````

````python
result == mammos_spindynamics.uppasd.Data("my-first-simulation")[0]
````

We get an object back, which allows us to inspect simulation output:

````python
result.Ms
````

````python
result.Cv
````

We can also access some simulation metadata, e.g. including the temperature value that we have selected and the runtime:

````python
result.info
````
