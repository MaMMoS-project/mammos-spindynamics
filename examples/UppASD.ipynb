# UppASD

```python
from mammos_spindynamics import uppasd
```

We start by defining a simulation object.
This object defines the location of the UppASD input files.

```python
sim = uppasd.Simulation(
    jfile="uppasd_data/jfile",
    momfile="uppasd_data/momfile",
    posfile="uppasd_data/posfile",
    inpsd="uppasd_data/inpsd-short.dat",
)
sim
```

## Single calculation: default `run` runner

We can run a single UppASD run specifying the temperature and other parameters we want to modify in `inpsd.dat`.

```python
sim.run(
    T=10,
    mcNstep=700,
)
```

In particular, we can see the values `temp 10` and `mcNstep 700` in the file `inpsd.dat`.

```python
with open("out/run_0/inpsd.dat") as f:
    print(f.read())
```

## Serial calculation: iterate default `run` runner

We can repeat different UppASD runs using different temperatures to generate different output directories.

```python
for T in [2, 5]:
    sim.run(
        T=T,
        mcNstep=700,
    )
```

## Serial calculation: `temparray` runner

We can use the `temparray` runner to run the simulation using different temperatures.

This will create a `run_temparray_<idx>` in the output directory (by default called `out`).
The index `idx` is the first non negative integer available.

```python
sim.run_temparray(
    T_array=[2, 5],
    mcNstep=700,
)
```

## Read result collection

We use the `ResultCollection` class to find all simulations run in a given output directory.

```python
res = uppasd.ResultCollection("out")
res
```
