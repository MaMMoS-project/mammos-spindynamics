# UppASD

```python
from mammos_spindynamics import uppasd
```

We start by defining a simulation object.

```python
sim_UppASD = uppasd.Simulation(
    jfile="uppasd_data/jfile",
    momfile="uppasd_data/momfile",
    posfile="uppasd_data/posfile",
    inpsd="uppasd_data/inpsd-short.dat",
)
sim_UppASD
```

Single UppASD run.

```python
sim_UppASD.run(
    T=10,
    mcNstep=700,
)
```

Run UppASD on an temperature array.

```python
for T in [2, 5]:
    sim_UppASD.run(
        T=T,
        mcNstep=700,
    )
```

```python
sim_UppASD.run_temparray(
    T_array=[2, 5],
    mcNstep=700,
)
```
